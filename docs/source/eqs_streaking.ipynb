# Attosecond Streaking (Definitions and Gradients)

## Definitions  

### Symbols and Indices

The SFA introduces a momentum axis `k`. In the numerical implementation a conjugate spatial axis `r` is also relevant. 
These two axis require new index convetions.  

$\quad q\quad\rightarrow\quad$ Spatial Domain  
$\quad b\quad\rightarrow\quad$ Momentum Domain  

Its a bit annoying that im already using $k$ and $p$ as indices. If its not an index $k$ refers to the momentum.

In addition the channel-dependent transition-dipole-matrix-element (DTME or $\rho^C$) and the channel dependent ionization potential $IP^C$ are introduced.  
In Streaking `A` always refers to the vectorpotential of the femtosecond pulse, while `E` refers to the field of the EUV-pulse.

### Streaking Amplitude (within Strong-Field-Approximation)  

For a single-channel the SFA is implemented as:

$S_{mb}^C = -i \cdot \sum\limits_k dt\cdot \rho_{bk}^C \cdot E_{mk} \cdot e^{-i t_k\cdot \left(IP^C + \frac{k_b^2}{2}\right)}\cdot e^{-i\phi_{bk}}$  

Where $S_{mb}$ is the streaking amplitude in the momenum domain, it can be fourier transform to the spatial domain to yield $S_{mq}$. The multi-channel SFA is simply a coherent sum over all channels, which translates to all gradients:

$S_{mb} = \sum\limits_C S_{mb}^C$  

$\frac{d S_{mb}}{dX} = \sum\limits_C \frac{d S_{mb}^C}{dX}$  


The vectorpotential-dependent part of the Volkov-Phase is:

$\phi_{bk} = \sum\limits_{j'=k}^{k_{\mathrm{max}}} dt\cdot \left(k_b A_{j'} + \frac{A_{j'}^2}{2}\right)$  

The DTME is momentum dependent and thus modulated by the vectorpotential. This can be numerically implemented using the Fourier-shift theorem.

$\rho_{bk}^C = \sum\limits_q D_{bq} \rho_q^C \cdot e^{i\cdot 2\pi r_q\cdot A_k}$  

## Gradients

### With respect to the Vectorpotential in frequency domain ($A_n$)  

Vectorpotential is required to be real for streaking. However through the chain-rule one can stil use wirtinger derivatives.  

$\frac{d S_{mb}^C}{d A_n} = \mathrm{FFT}\{\frac{d S_{mb}^C}{d A_j}\}$  

$\frac{d S_{mb}^C}{d A_j} = \frac{d \Re\{A_j\}}{d A_j} \frac{d S_{mb}^C}{d \Re\{A_j\}} = \frac{1}{2} \frac{d S_{mb}^C}{d \Re\{A_j\}} = \frac{d S_{mb}^C}{d A_j^*}$  

In the following the gradient is taken with respect to the real-part of $A$, for ease on the eyes the $\Re\{\}$ is omitted.

$\frac{d S_{mb}^C}{d A_j} = -i\cdot\sum\limits_k dt \cdot E_{mk} \cdot e^{-i t_k\cdot \left(IP^C + \frac{k_b^2}{2}\right)} \cdot \left(-i\cdot \rho_{bk}^C\cdot e^{-i\phi_{bk}}\cdot \frac{d \phi_{bk}}{d A_j} + e^{-i\phi_{bk}}\cdot\frac{d\rho_{bk}^C}{d A_j} \right)$  

In the case-definition $k$ is the index and not the momentum, the momentum is $k_b$. Can be implemented with binary mask.

$\frac{d \phi_{bk}}{d A_j} = \begin{cases}dt \cdot (k_b + A_j) & \text{if } k \le j \le k_{max} \\
0 & \text{else}\end{cases} $

$\frac{d \rho_{bk}^C}{d A_j} = \sum\limits_q D_{bq}\rho_q^C\cdot e^{i\cdot 2\pi r_q\cdot A_k}\cdot i\cdot 2\pi r_q\cdot \delta_{kj}$

### With respect to the attosecond pulse in frequency domain $\left(E_n\right)$  

$\frac{d S_{mb}^C}{d E_n} = -i \cdot \sum\limits_k dt\cdot \rho_{bk}^C \cdot e^{-i t_k\cdot \left(IP^C + \frac{k_b^2}{2}\right)}\cdot e^{-i\phi_{bk}} \cdot D_{kn} e^{-i\omega_n\tau_m}$  

### With respect to the dipole transition matrix element in momentum domain ($\rho_b$)  

$\frac{d S_{mb}^C}{d \rho_q^{C'}} = -i \cdot \sum\limits_k dt\cdot E_{mk} \cdot e^{-i t_k\cdot \left(IP^C + \frac{k_b^2}{2}\right)}\cdot e^{-i\phi_{bk}} \cdot D_{bq}\cdot e^{i\cdot 2\pi r_q A_k} \cdot \delta_{CC'}$  

$\frac{d S_{mb}^C}{d \rho_b^{C'}} = \mathrm{FFT}_{qb}\{\frac{d S_{mb}^C}{d \rho_q^{C'}}\}$
